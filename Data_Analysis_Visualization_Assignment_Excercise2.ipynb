```python
# Excercise 2: World Happiness Data Exploration

import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
# 1. Setup and data loading

df = pd.read_csv('2019.csv')
df.head(7)
```

```python
# 3. Happiness score distribution: Analyse the distribution of happiness scores accross countries.

sns.histplot(df['Score'], bins=5, kde=True)
plt.pyplot.title('Distribution of Happiness Scores (2019)')
plt.pyplot.xlabel('Happiness Score')
plt.pyplot.ylabel('# of Countries')
plt.pyplot.show()

# we can observe from the histoogram plt below that about mostof the countries have happiness score ranginf from 4 to 6.
# This appears to be cocentrated in the center
```

```python
# 4. Top and Bottom Countries: Identify the happiest andleast happy countries.

# Sorting dataframe on 'Score' column and display top 10 happiest countries.

df.sort_values('Score', ascending=True)
df.head(10)
```

```python
# 4. Continue...
# display bottom least happy countries
df.tail(10)
```

```python
# 5. Visualize top 10 happiest countries: Create a chart to showcase the happiest countries.
top_10_Happy = df.sort_values('Score', ascending = False).head(10)

sns.barplot(x='Score', y='Country or region', data=top_10_Happy)
plt.pyplot.title('Top 10 Happiest Countries or Region in 2019')
plt.pyplot.xlabel('Score')
plt.pyplot.ylabel('Country or Region')
plt.pyplot.show()
```

```python
# 6. Happiness vs. GDP per Capita: Investigate the relationship between economic performance and happiness.

sns.scatterplot(x='GDP per capita', y='Score', data=df)
plt.pyplot.title('Happiness Score vs. GDP per Capita (2019)')
plt.pyplot.xlabel('GDP per Capita')
plt.pyplot.ylabel('Happiness Score')
plt.pyplot.show()

# from the scatter plot below we can observe a clear relationship between 'GDP per Capita' and 'Happiness'. We can see that happiness
# increases with increase in per capita GDP. There are a few exception countries where happiness does not appear to increase with
# increase per capita GDP. However, majority shows a positive trend.
```

```python
correlation = df['Score'].corr(df['GDP per capita'])
print(f"The correlation between Happiness Score and GDP per capita is: {correlation:.2f}")

# Corelation between Happiness and per capita GDP is 0.79, which is positive and confirms our earlier statement
# that happiness increases with increaase GDP.
```

```python
# 7. Additional Analysis (Optional Extension): verify relationship between Healthy life expectancy and Score  Freedom to make life choices

sns.scatterplot(x='Healthy life expectancy', y='Score', data=df)
plt.pyplot.title('Happiness Score vs. Healthy life expectancy (2019)')
plt.pyplot.xlabel('Healthy life expectancy')
plt.pyplot.ylabel('Happiness Score')
plt.pyplot.show()

# from the scatter plot below, we can observe a positive relationship between healthy life expenctancy and happiness score.
# With increase in healty life expectancy the people appears to be more happy.
```

```python
# 7. Additional Analysis (Optional Extension): verify relationship between Freedom to make life choices and Score.

sns.scatterplot(x='Freedom to make life choices', y='Score', data=df)
plt.pyplot.title('Happiness Score vs. Freedom to make life choices (2019)')
plt.pyplot.xlabel('Freedom to make life choices')
plt.pyplot.ylabel('Happiness Score')
plt.pyplot.show()

# from the scatter plot below, we can observe a positive relationship between Freedom to make life choices and happiness score.
# However, the distribution is more scattered so there is still a variablility suggeting other factors could influence happiness.
```
